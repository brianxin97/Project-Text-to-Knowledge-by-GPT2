```python
import os

def count_paragraphs_in_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        paragraphs = content.split('\n\n')
        return sum(1 for p in paragraphs if p.strip())

def count_paragraphs_in_folder(folder_path):
    total_paragraphs = 0
    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            file_path = os.path.join(folder_path, filename)
            total_paragraphs += count_paragraphs_in_file(file_path)
    return total_paragraphs

folder_path = '/hkfs/home/project/hk-project-p00201316/st_st176945/testout'
print(f"Total number of paragraphs across all text files in the folder: {count_paragraphs_in_folder(folder_path)}")
```
