```python
import json
import os
from tqdm import tqdm

input_file_path = '/hkfs/home/project/hk-project-p00201316/st_st176945/entity_cache.json'
output_file_path = '/hkfs/home/project/hk-project-p00201316/st_st176945/new_para_cache/para_cache.json'

if not os.path.exists(input_file_path):
    print("file doesn't exist")
    exit()

with open(input_file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

progress_bar = tqdm(total=len(data), desc="Processing", unit="entry")

processed_data = {}
for key in data:
    lines = data[key].split('\n')
   
    processed_data[key] = ' '.join(lines).strip()
    
    progress_bar.update(1)

progress_bar.close()

with open(output_file_path, 'w', encoding='utf-8') as file:
    json.dump(processed_data, file, ensure_ascii=False, indent=4)

print(f"finish, save to {output_file_path} ")
```
