```python
import json

file_path = '/hkfs/home/project/hk-project-p00201316/st_st176945/testin/1000_reformed.jsonl'

empty_count = 0
no_description_found_count = 0

with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        data = json.loads(line)
        paragraph = data.get("paragraph", "")
        if "]: " in paragraph:
            description_part = paragraph.split("]:", 1)[1].strip()
            if not description_part:  
                empty_count += 1
            if description_part == "No description found": 
                no_description_found_count += 1

print("Empty lines count:", empty_count)
print("Lines with 'No description found':", no_description_found_count)
```
