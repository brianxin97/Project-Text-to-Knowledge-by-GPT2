```python
import os

def count_lines_in_jsonl_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return sum(1 for line in file)

def count_total_lines_in_directory(directory_path):
    total_lines = 0
    for filename in os.listdir(directory_path):
        if filename.endswith('.jsonl'):
            file_path = os.path.join(directory_path, filename)
            total_lines += count_lines_in_jsonl_file(file_path)
    return total_lines

directory_path = '/hkfs/home/project/hk-project-p00201316/st_st176945/entity_rels_reformed'

total_lines = count_total_lines_in_directory(directory_path)
print(f'Total lines in all .jsonl files: {total_lines}')
```
